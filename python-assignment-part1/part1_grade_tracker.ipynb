```python
# Task 1: Data Parsing & Profile Cleaning

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

# we will also store cleaned data if needed later
cleaned_students = []

for student in raw_students:

    #remove spaces and convert to title case
    name = student["name"].strip().title()

    # converting roll number to integer
    roll = int(student["roll"])

    # converting marks string into list of integers
    marks_list = student["marks_str"].split(", ")
    marks = []

    for m in marks_list:
        marks.append(int(m))

    # checking if name is valid
    words = name.split()
    valid = True

    for w in words:
        if not w.isalpha():
            valid = False
            break

    # printing validation result
    if valid:
        print(f"{name} → ✓ Valid name")
    else:
        print(f"{name} → ✗ Invalid name")

    # printing formatted profile card
    print("================================")
    print(f"Student : {name}")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("================================\n")

    # storing cleaned student
    cleaned_students.append({
        "name": name,
        "roll": roll,
        "marks": marks
    })


# finding student with roll number 103
for student in cleaned_students:
    if student["roll"] == 103:
        print("Special Output for Roll 103:")
        print("UPPERCASE :", student["name"].upper())
        print("lowercase :", student["name"].lower())
```

```python
# Task 2: Marks Analysis Using Loops & Conditionals

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

print("Student:", student_name)
print("\nSubject-wise Performance:\n")

# using for loop to print subject, marks and grade
for i in range(len(subjects)):
    m = marks[i]

    # deciding grade based on marks
    if m >= 90:
        grade = "A+"
    elif m >= 80:
        grade = "A"
    elif m >= 70:
        grade = "B"
    elif m >= 60:
        grade = "C"
    else:
        grade = "F"

    print(f"{subjects[i]} : {m} → Grade {grade}")


# calculating total and average
total = sum(marks)
average = total / len(marks)

print("\nTotal Marks :", total)
print("Average Marks :", round(average, 2))


# finding highest and lowest scoring subject
max_marks = max(marks)
min_marks = min(marks)

# finding corresponding subject names
for i in range(len(marks)):
    if marks[i] == max_marks:
        highest_subject = subjects[i]
    if marks[i] == min_marks:
        lowest_subject = subjects[i]

print("\nHighest Scoring Subject :", highest_subject, "-", max_marks)
print("Lowest Scoring Subject :", lowest_subject, "-", min_marks)


# while loop for adding new subjects

print("\n--- Add New Subjects (type 'done' to stop) ---")

new_count = 0

while True:
    sub = input("Enter subject name: ")

    if sub.lower() == "done":
        break

    mark_input = input("Enter marks (0-100): ")

    # checking if input is numeric
    if not mark_input.isdigit():
        print("Invalid input! Marks should be a number.")
        continue

    mark = int(mark_input)

    # checking valid range
    if mark < 0 or mark > 100:
        print("Invalid marks! Enter between 0 and 100.")
        continue

    # if valid, add to list
    subjects.append(sub)
    marks.append(mark)
    new_count += 1


# after loop ends
print("\nNew subjects added:", new_count)

updated_total = sum(marks)
updated_avg = updated_total / len(marks)

print("Updated Average:", round(updated_avg, 2))
```

```python
# Task 3: Class Performance Summary

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print("Name              | Average | Status")
print("----------------------------------------")

pass_count = 0
fail_count = 0

topper_name = ""
topper_avg = 0

total_avg_sum = 0   # to calculate class average later

# looping through each student
for student in class_data:
    name = student[0]
    marks = student[1]

    avg = sum(marks) / len(marks)
    avg = round(avg, 2)

    total_avg_sum += avg

    # deciding pass/fail
    if avg >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    # checking topper
    if avg > topper_avg:
        topper_avg = avg
        topper_name = name

    # printing formatted row
    print(f"{name:<18} | {avg:>7.2f} | {status}")


# after loop
print("\nStudents Passed:", pass_count)
print("Students Failed:", fail_count)

print("Class Topper:", topper_name, "-", topper_avg)

class_average = total_avg_sum / len(class_data)
print("Class Average:", round(class_average, 2))
```

```python
# Task 4: String Manipulation Utility

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# Step 1
# removing extra spaces from beginning and end
clean_essay = essay.strip()

print("Step 1 - Clean Essay:")
print(clean_essay)
print()
```

```python
# Step 2
# converting cleaned essay into title case
title_case = clean_essay.title()

print("Step 2 - Title Case:")
print(title_case)
print()
```

```python
# Step 3
# counting how many times 'python' appears
count_python = clean_essay.count("python")

print("Step 3 - Count of 'python':", count_python)
print()
```

```python
# Step 4
# replacing 'python' with 'Python 🐍'
replaced_text = clean_essay.replace("python", "Python 🐍")

print("Step 4 - After Replacement:")
print(replaced_text)
print()
```

```python
# Step 5
# splitting essay into sentences
sentences = clean_essay.split(". ")

print("Step 5 - List of Sentences:")
print(sentences)
print()
```

```python
# Step 6
# printing each sentence with numbering

print("Step 6 - Numbered Sentences:")

i = 1
for s in sentences:

    # adding period if missing
    if not s.endswith("."):
        s = s + "."

    print(f"{i}. {s}")
    i += 1
```
